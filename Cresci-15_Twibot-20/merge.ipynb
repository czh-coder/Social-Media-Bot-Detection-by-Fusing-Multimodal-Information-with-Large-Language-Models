```python
from typing import List, Tuple,Dict

class BotDataset:
    def __init__(self, 
                 cat_properties: torch.Tensor,    # 类别属性张量
                 num_properties: torch.Tensor,    # 数值属性张量
                 des_tensor: torch.Tensor,        # 描述张量
                 tweets_tensor: torch.Tensor,     # 推文张量
                 edge_index: torch.Tensor,        # 边索引 (2, E)
                 edge_type: torch.Tensor,         # 边类型 (E,)
                 train_idx: torch.Tensor,         # 训练索引：一维Tensor
                 val_idx: torch.Tensor,           # 验证索引：一维Tensor
                 test_idx: torch.Tensor,          # 测试索引：一维Tensor
                 label: torch.Tensor,             # 标签: 一维Tensor
                 Shb_prompt: Dict[int, str],      # JSON格式字典：key为序号，value为字符串
                 Stx_prompt: Dict[int, str]):     # JSON格式字典：key为序号，value为字符串
        self.cat_properties = cat_properties
        self.num_properties = num_properties
        self.des_tensor = des_tensor
        self.tweets_tensor = tweets_tensor
        self.edge_index = edge_index
        self.edge_type = edge_type
        self.train_idx = train_idx
        self.val_idx = val_idx
        self.test_idx = test_idx
        self.label = label
        self.Shb_prompt = Shb_prompt
        self.Stx_prompt = Stx_prompt
        
        # 节点数（假设以cat_properties的第一维作为节点数）
        self.num_nodes = cat_properties.shape[0]
        # prompt字典的最大key（用于后续偏移）
        self.shb_max_key = max(Shb_prompt.keys()) if Shb_prompt else -1
        self.stx_max_key = max(Stx_prompt.keys()) if Stx_prompt else -1
```

```python
def merge_datasets(ds1: BotDataset, ds2: BotDataset) -> BotDataset:
    """
    合并两个BotDataset数据集（含prompt字典合并）
    :param ds1: 第一个数据集（Twibot20）
    :param ds2: 第二个数据集（Cresci15）
    :return: 合并后的数据集
    """
    # -------------------- 2.1 拼接张量数据 --------------------
    merged_cat = torch.cat([ds1.cat_properties, ds2.cat_properties], dim=0)
    merged_num = torch.cat([ds1.num_properties, ds2.num_properties], dim=0)
    merged_des = torch.cat([ds1.des_tensor, ds2.des_tensor], dim=0)
    merged_tweets = torch.cat([ds1.tweets_tensor, ds2.tweets_tensor], dim=0)
    
    # 拼接边索引和边类型（ds2的边索引需偏移节点数）
    ds2_edge_index = ds2.edge_index + ds1.num_nodes
    merged_edge_index = torch.cat([ds1.edge_index, ds2_edge_index], dim=1)
    merged_edge_type = torch.cat([ds1.edge_type, ds2.edge_type], dim=0)

    # 拼接标签
    merged_label = torch.cat([ds1.label, ds2.label], dim=0)
    
    # -------------------- 2.2 处理Tensor索引的偏移与合并 --------------------
    offset = ds1.num_nodes  # 节点索引偏移量：ds1的节点数
    ds2_train = ds2.train_idx + offset
    ds2_val = ds2.val_idx + offset
    ds2_test = ds2.test_idx + offset
    
    merged_train = torch.cat([ds1.train_idx, ds2_train], dim=0)
    merged_val = torch.cat([ds1.val_idx, ds2_val], dim=0)
    merged_test = torch.cat([ds1.test_idx, ds2_test], dim=0)
    
    # -------------------- 2.3 合并prompt字典（核心：key偏移） --------------------
    def merge_prompt_dict(prompt1: Dict[int, str], prompt2: Dict[int, str]) -> Dict[int, str]:
        """
        合并两个prompt字典，对第二个字典的key进行偏移
        :param prompt1: Twibot20的prompt字典
        :param prompt2: Cresci15的prompt字典
        :return: 合并后的prompt字典
        """
        if not prompt1 and not prompt2:
            return {}
        # 计算偏移量：prompt1的最大key + 1（若prompt1为空，偏移量为0）
        offset_key = max(prompt1.keys()) + 1 if prompt1 else 0
        # 对prompt2的key进行偏移，生成新字典
        prompt2_shifted = {k + offset_key: v for k, v in prompt2.items()}
        # 合并两个字典（prompt1在前，prompt2_shifted在后）
        merged_prompt = {**prompt1, **prompt2_shifted}
        return merged_prompt
    
    # 合并Shb_prompt和Stx_prompt
    merged_Shb = merge_prompt_dict(ds1.Shb_prompt, ds2.Shb_prompt)
    merged_Stx = merge_prompt_dict(ds1.Stx_prompt, ds2.Stx_prompt)
    
    # -------------------- 2.4 创建合并后的数据集 --------------------
    merged_ds = BotDataset(
        cat_properties=merged_cat,
        num_properties=merged_num,
        des_tensor=merged_des,
        tweets_tensor=merged_tweets,
        edge_index=merged_edge_index,
        edge_type=merged_edge_type,
        train_idx=merged_train,
        val_idx=merged_val,
        test_idx=merged_test,
        label=merged_label,
        Shb_prompt=merged_Shb,
        Stx_prompt=merged_Stx
    )
    
    return merged_ds
```
